##Eliminaci√≥n de At√≠picos

```python
columnas_continuas = ["Relative_Compactness", "Surface_Area", "Wall_Area","Roof_Area","Overall_Height"]
variables_continuas = eficiencia_df[columnas_continuas]
variables_continuas
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(variables_continuas)
scaled_data = pd.DataFrame(scaled_data, columns=[columnas_continuas])
scaled_data
```

```python
columnas = set(["Relative_Compactness", "Surface_Area", "Wall_Area", "Roof_Area", "Overall_Height", "Orientation", "Glazing_Area", "Glazing_Area_Distribution", "Heating_Load", "Cooling_Load"])
columnas_continuas = set(["Relative_Compactness", "Surface_Area", "Wall_Area","Roof_Area","Overall_Height", "Cooling_Load"])
cols = list(columnas - columnas_continuas)
Data_Final = pd.concat([scaled_data, eficiencia_df[cols]], axis=1)
Data_Final
```

```python
import matplotlib.pyplot as plt

#plt.figure(figsize=(18,9))
#scaled_data.boxplot()
fig , ax1 = plt.subplots(figsize=(10,6))
ax1.boxplot(scaled_data)
ax1.set_xticklabels(columnas_continuas, rotation=90)
plt.title("Variables n√∫mericas del dataset", fontsize=20)
plt.show()
```

**De acuerdo al resultado de la gr√°fica. No se encuentran valores at√≠picos**



```python
Q1 = scaled_data.quantile(0.25)
Q3 = scaled_data.quantile(0.75)
IQR = Q3 - Q1

datos_tipicos = scaled_data[~((scaled_data < (Q1 - 1.5 * IQR)) |(scaled_data > (Q3 + 1.5 * IQR))).any(axis=1)]
datos_tipicos.shape
```

**De acuerdo con el an√°lisis del rango intercuartil, no se encuentran datos at√≠picos**

##Relaci√≥n entre Variables

```python
import seaborn as sns
correlation_matrix = variables_continuas.corr()

plt.figure(figsize=(10, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```

**Vemos que la correlaci√≥n entre Heating Load y Cooling Load es muy fuerte. As√≠ mismo, la correlaci√≥n entre Surface Area y Relative Compactness tambi√©n es muy fuerte, pero en este caso es inversa. La correlaci√≥n de Surafce Area tanto con Heating Load como con Cooling Load es fuerte e inversa. En general se observa tambi√©n que las variables Heating Load y Cooling Load tienen una correlaci√≥n muy similar con las dem√°s variables.**

```python
sns.pairplot(variables_continuas)
```

**Con las gr√°ficas para el an√°lisis bivariante, podemos ver c√≥mo se comportan los datos en relaci√≥n a los dem√°s. Se corrobora que solo las variables Heating Load y Cooling Load, tienen valores continuios y una correlaci√≥n fuerte. Las dem√°s variables tienen valores discretos que no aplican para el an√°lisis, por lo tanto se toman como variables categoricas.
 Vemos que tanto las variables Heating Load como Cooling Load tienen una distribuci√≥n bimodal.**
